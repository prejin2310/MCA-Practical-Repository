```python
#line plot 
import matplotlib.pyplot as plt
y=[3,8,1,10]
plt.plot(y,color='green') #linestyle='--'
plt.show()
```

```python
#box plot with random data
import numpy as np
import matplotlib.pyplot as plt
data=np.random.rand(100) #generate some random variables
plt.boxplot(data)
plt.show()
```

```python
#box plot 
import matplotlib.pyplot as plt
data=[10,25,32,56,78,85,100]
plt.boxplot(data)
plt.show()
```

```python
#barchart 
import matplotlib.pyplot as plt
data=[75,25,50,200]
ind=[1,2,3,4]
plt.bar(ind,data) #use barh - for horizondal view
plt.show()
```

```python
#Scatterplot
x=[5,7,8,7,2,17,2,9,4,11,12,9,6]
y=[99,86,87,88,111,86,103,87,94,78,77,85,86]
plt.scatter(x,y)
plt.show()
```

```python
#Histogram - display number of occurance
import numpy as np
import matplotlib.pyplot as plt
height=np.arange(150,200,2)
height
plt.hist(height,rwidth=.9)
```

```python
import numpy as np
import matplotlib.pyplot as plt
height=np.arange(150,200,2)
height
bins=[150,165,175,190,200]
plt.hist(height,rwidth=.9,bins=bins,color='yellow')
```
